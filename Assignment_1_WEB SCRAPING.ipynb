```python
from bs4 import BeautifulSoup
import requests
```

```python
# 1. Write a python program to display all the header tags from wikipedia.org.

url=('https://www.wikipedia.org/')
source=requests.get(url)
print(source)

```

```python
soup=BeautifulSoup(source.content,'html.parser')
soup
```

```python
titles = soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
print('List all the header tags :', *titles, sep='\n\n')
```

```python
url=('https://www.imdb.com/chart/top/')
source= requests.get(url)
print(source)
```

```python
soup=BeautifulSoup(source.content,'html.parser')
soup
```

```python
movies= soup.find('tbody',class_="lister-list")
movies= soup.find('tbody',class_="lister-list").find_all('tr')
print(len(movies))
```

```python
for movie in movies:
        Name=movie.find('td',class_="titleColumn")
        #print(Name)
        Name=movie.find('td',class_="titleColumn").a.text
        #print(Name)
    
        Year_of_release= movie.find('td',class_="titleColumn").span.text
        Year_of_release= movie.find('td',class_="titleColumn").span.text.strip('()')
        #print(Year_of_release)
        
        Rating=movie.find('td',class_="ratingColumn imdbRating")
        Rating=movie.find('td',class_="ratingColumn imdbRating").strong.text
        #print(Rating)
        
        print(Name,Rating,Year_of_release)
```

```python
# Making dataframe
import pandas as pd
df=pd.DataFrame({'name':Name,'rating':Rating,'year':Year_of_release})
```

```python
url=('https://meesho.com/bags-ladies/pl/p7vbp')
source= requests.get(url)
print(source)
```

```python
soup=BeautifulSoup(source.content,'html.parser')
soup
```

```python
product_name=soup.find('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS")
product_name.text

product_price=soup.find('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy")
product_price.text

product_discounts=soup.find('span',class_="Text__StyledText-sc-oo0kvp-0 lnonyH")
product_discounts.text
```

```python
# Srapping Multiple using find_all

product_name=[]
for i in soup.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
    product_name.append(i.text)
    
product_name

product_price=[]
for i in soup.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    product_price.append(i.text)
    
product_price


product_discounts=[]
for i in soup.find_all('span',class_="Text__StyledText-sc-oo0kvp-0 lnonyH"):
    product_discounts.append(i.text)
    
product_discounts



```

```python
print(product_name,product_price,product_discounts)
```

```python
# Making dataframe
import pandas as pd
df=pd.DataFrame({'Name':product_name,'Price':product_price,'Discounts':product_discounts})
df
```
